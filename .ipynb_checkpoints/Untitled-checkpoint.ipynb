```python
df['activity'] = df['activity'].str.split(', ')
```

```python
countries = pd.read_json('data/slim-3.json')
```

```python
code, country_Name = countries['alpha-3'].tolist(), countries['name']
```

```python
c_dict= dict(zip(code,country_Name))
```

```python
activity_country = df \
.explode('activity') \
.groupby(['activity','country_code']) \
.size().reset_index().rename(columns={0:'value'})
```

```python
activity_country['country_code'] = activity_country['country_code'].map(c_dict)
```

```python
activity_country.head()
```

```python
%%time
country_act_df = activity_country.pivot(index='activity', values='value', columns='country_code').fillna(0).reset_index()
```

```python
country_act_df.rename(columns={'activity':'Name'}, inplace=True)
```

```python
activity_country.head()
```

```python
%%time
country_act_df = activity_country.pivot(index='activity', values='value', columns='country_code').fillna(0).reset_index()
```

```python
country_act_df.to_csv('data/my_all_countries.csv', quoting=0)
```

```python
activity_encounter = df \
.explode('activity') \
.groupby(['activity','encounter_type']) \
.size().reset_index().rename(columns={0:'value'})
```

```python
%%time
activity_encounter_df = activity_encounter.pivot(index='activity', values='value', columns='encounter_type').fillna(0).reset_index()
```

```python
activity_encounter_df.rename(columns={'activity':'Name'}, inplace=True)
```

```python
activity_encounter_df.to_csv('data/my_all_encounter.csv')
```
